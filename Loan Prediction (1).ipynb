```python
X = df_train.drop('Loan_ID', axis=1)
X = df.drop('Loan_Status', axis=1)
y = df['Loan_Status']
X.fillna( method ='ffill', inplace = True)
X = pd.get_dummies(X , drop_first = True)
X
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(X, y , test_size = 0.25 , random_state = 42, stratify=y)
```

```python
y_test.value_counts()
```

```python
y_train.value_counts()
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train_scale = sc.fit_transform(x_train)
x_test_scale = sc.transform(x_test)
```

```python
x_train_scale
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train_scale,y_train)
y_pred = lr.predict(x_test_scale)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
from sklearn.svm import SVC
classifier = SVC(kernel="rbf")
classifier.fit(x_train_scale,y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred = classifier.predict(x_test_scale)
accuracy_score(y_test, y_pred)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(x_train_scale,y_train)
y_pred = rf.predict(x_test_scale)
print('evaluation on training', rfc.score(x_train_scale,y_train))
print('evaluation on testing', rfc.score(x_test_scale,y_test))
```
